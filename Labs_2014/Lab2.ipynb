###Choosing a model

The exponential distribution occurs naturally when describing the lengths of the inter-arrival times in a homogeneous Poisson process.

It takes the form:
$$
f(x;\lambda) = \begin{cases}
\lambda e^{-\lambda x} & x \ge 0, \\
0 & x < 0.
\end{cases}
$$

From Wikipedia: In probability theory, a Poisson process is a stochastic process which counts the number of events and the time that these events occur in a given time interval. The time between each pair of consecutive events has an exponential distribution with parameter $\lambda$ and each of these inter-arrival times is assumed to be independent of other inter-arrival times. The process is named after the French mathematician Siméon Denis Poisson and is a good model of radioactive decay, telephone calls and requests for a particular document on a web server, among many other phenomena.

In our example above, we have the arrival times of the babies. There is no reason to expect any specific clustering in time, so one could think of modelling the arrival of the babies via a poisson process.

Furthermore, the Poisson distribution can be used to model the number of births each hour over the 24-hour period.

```python
f = lambda x, l: l*np.exp(-l*x)*(x>0)
xpts=np.arange(-2,3,0.1)
plt.plot(xpts,f(xpts, 2),'o');
plt.xlabel("x")
plt.ylabel("exponential pdf")
```

###How would we draw from this distribution?

We can use the usual inverse transformation.

The CDF of the exponential distribution is:

$$
F(x,\lambda) = 1-e^{-\lambda x}
$$

Given a random variate U drawn from the uniform distribution on the unit interval, we can thus generate samples from the exponential distribution as:

$$
x = -\frac{ln(1-U)}{\lambda} = -\frac{ln(U)}{\lambda}
$$

This was done in class, so I wont repeat it here. Lets use the built in machinery in `scipy.stats` instead:

```python
from scipy.stats import expon
plt.plot(xpts,expon.pdf(xpts, scale=1./2.),'o')
plt.hist(expon.rvs(size=1000, scale=1./2.), normed=True, alpha=0.5, bins=30);
plt.xlabel("x")
plt.title("exponential pdf and samples(normalized)");
```

In `scipy.stats`, you can alternatively create a frozen object, which holds values of things like the scale

```python
rv = expon(scale=0.5)
plt.plot(xpts,rv.pdf(xpts),'o')
plt.hist(rv.rvs(size=1000), normed=True, alpha=0.5, bins=30);
plt.plot(xpts, rv.cdf(xpts));
plt.xlabel("x")
plt.title("exponential pdf, cdf and samples(normalized)");
```

The mean or of an exponentially distributed random variable X with rate parameter $\lambda$ can be analytically calculated as

$$\Ex[X] = \frac{1}{\lambda}.$$

This makes intuitive sense: if you get babies at an average rate of 2 per hour, then you can expect to wait half an hour for every baby.

The variance of X is given by

$$\Var[X] = \frac{1}{\lambda^2}.$$

so the standard deviatiation is equal to the mean, just as in the discrete Poisson distribution.

```python
lambda_from_mean = 1./timediffs.mean()
print lambda_from_mean, 1./lambda_from_mean
```

## 2. FREQUENTIST STATISTICS 

(some text taken from Murphy)

In frequentist statistics, a parameter estimate is computed by applying an estimator $F$ to some data $D$, so $\est{\lambda} = F(D)$. 

**The parameter is viewed as fixed and the data as random, which is the exact opposite of the Bayesian approach. **

In other words, on the assumption that an exponential distribution is a good description of the baby arrival process, there is some true $\trueval{\lambda}$ which defines it. We dont know this. The best we can do to start with is to estimate  a lambda from the data set we have, which we denote $\est{\lambda}$. Now if we had many replications of this data set: that is, data from other days, an **ensemble** of data sets, for example, we can compute other $\est{\lambda}$, and begin to construct the "sampling distribution" of $\lambda$.

The uncertainty in the parameter estimate can be measured by computing the sampling distribution of the estimator. 
What you are doing is sampling many Data Sets $D_i$, say M of them, each of size N, from some true model $p(\cdot|\trueval{\lambda})$. We will now calculate M $\est{\lambda}_i$, one for each dataset. As we let $M \rightarrow \infty$, the distribution induced on $\est{\lambda}$ is the sampling distribution of the estimator.

### Maximum Likelyhood Estimation

One of the techniques used to estimate such parameters in frequentist statistics is **maximum likelyhood estimation**. This will be talked more about in class this coming week, but briefly, the idea behind it is:

The product 

$$
L(\lambda) = \prod_{i=1}^n P(x_i | \lambda)
$$

gives us a measure of how likely it is to observe values $x_1,...,x_n$ given the parameters $\lambda$. Maximum likelihood fitting consists of choosing the appropriate "likelyhood" function $L=P(X|\lambda) to maximize for a given set of observations. How likely are the observations if the model is true?

Often it is easier and numerically more stable to maximise the log likelyhood:

$$
\ell(\lambda) = \sum_{i=1}^n ln(P(x_i | \lambda))
$$

In the case of the exponential distribution we have:

$$
\ell(lambda) = \sum_{i=1}^n ln(\lambda e^{-\lambda x_i}) = \sum_{i=1}^n \left( ln(\lambda) - \lambda x_i \right).
$$

Maximizing this:

$$
\frac{d \ell}{d\lambda} = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0
$$

and thus:

$$
\est{\lambda_{MLE}} = \frac{1}{n}\sum_{i=1}^n x_i,
$$

which is identical to the simple estimator we used above. Usually one is not so lucky and one must use numerical optimization techniques.

A crucial property is that, for many commonly occurring situations, maximum likelihood parameter estimators have an approximate normal distribution when n is large. 

### Inference 

Just having an estimate is no good. We will want to put confidence intervals on the estimation of the parameters. This presents a conundrum: we have access to only one sample, but want to compute a error estimate over multiple samples, using an estimator such as the standard deviation.

If the model we use is a simple one like the gaussian, we can just use the properties of the gaussian to calculate confidence intervals. For instance, we want to calculate the area under the curve, such that the estimator in 95% of the samples would be contained in it. (One could, pathologically have our dataset's estimator outside this interval, but then one would argue that our model isnt very good!)

But many a time our likelyhoods and distributions arent symmetric, and analytically tractable. In the last two decades, **resampling** the ONE dataset we have has become computationally feasible. Resampling involves making new samples from the observations, "each of which can be analyzed in the same way to see how the statistics depend on plausible random variations in the observations"[REF]. One way to do this is the Bootstrap. 

### Bootstrap

(again from murphy)

The bootstrap is a simple Monte Carlo technique to approximate the sampling distribution.  If we knew the true parameters $\trueval{\theta}$, we could generate many (say M) fake datasets, each of size N, from the true distribution,  $p(\cdot|\trueval{\theta})$. We could then compute our estimator from each sample, $\est{\theta} = F(X_m)$ and use the empirical distribution of the resulting samples as our estimate of the sampling distribution. Since $\theta$ is unknown, the idea of the parametric bootstrap is to generate the samples using the $\est{\theta}$ estimated from D instead. An alternative, called the non-parametric bootstrap, is to sample with replacement the X from the empirical distribution of the original data D, and then compute the induced distribution as before. 

Babu and Singh showed you need roughly $N \sim n(ln(n))^2$ samples. 

## MONTE CARLO

We can use Monte Carlo to approximate the expected value of any function of a random variable. Draw samples, and then compute the mean of the function applied to the samples. This can be written as follows:

$$
\Ex[f(X)] = \int f(x) p(x) dx \approx \SampleAvg f(x_s)
$$
where $x_s \sim p(X)$. 
By varying the function $f$, we can approximate many quantities of interest, such as:


* $\avg{x} = \SampleAvg x_s \rightarrow \Ex[X]$, the mean
* $\SampleAvg (x_s - \avg{x})^2 \rightarrow \Var[X]$, the variance
* $\SampleAvg \indic(x_s \le c) \rightarrow P(X \le c)$, the cdf

For the last one, you could just count the samples. Similarly, once you have the samples, you can find the median by getting the median of the samples.

### Where is Monte Carlo used?

So far, we have used it in a rather simple way: just simple simulations to draw, and simple calculation of expectations over the bootstrap. But yes, its that fundamental. However, it comes into its own in bayesian statistics, where we integrate, marginalize, 'expect', and normalize intractable 'posterior' probability distributions over their parameters (as opposed to simulated or real ensembles of data sets).. The ways to generate samples remain the inverse, rejection, importance, and a new one we will learn, MCMC, and then we compute these expectations and integrals using the samples.

The exponential distribution is a special case of the gamma distribution, which may be used when a positive continuous variable is desired. Specifically,
its $\Gam{(1, \lambda)}$. The Gamma function in general is:

$$\Gam{(x;\alpha,\beta)} = \frac{\beta^{\alpha} x^{\alpha-1} e^{-\beta x}}{\Gamma(\alpha)} \quad \text{ for } x \geq 0 \text{ and } \alpha, \beta > 0$$

where $\alpha$ is called the shape parameter, and $\beta$ is the rate parameter. The distribution is used as a conjugate prior to many others such as exponential, gaussian, and poisson. Thus the effort expended in understanding it will be useful later. It has been used to model rainfall.

In this parametrization, the mean of the gamma distribution is 
$$\Ex[X] = \frac{\alpha}{\beta}.$$

```python
from scipy.stats import gamma
xpos=np.arange(0,5,0.1)
def gammafunc(x,alpha, beta):
    return gamma.pdf(x,alpha, scale=1./beta)
plt.plot(xpos,gammafunc(xpos, 1.,2.))
plt.plot(xpos,gammafunc(xpos, 2.,2.))
plt.plot(xpos,gammafunc(xpos, 2.,1.))

plt.xlabel("x")
plt.title("gamma pdfs");
```

### Rejection Method for samples

Now say we wanted to be able to draw from this distribution, for some reason. 
How are random variates generated for the gamma function? Indeed, when you choose a `scipy.stats gamma.rvs()` function, what is going on under the hood? With the advent of MCMC, it is important to have fast algorithms to generate random samples.

One way to do this, is the rejection method. However, as you know, if the function you wish to sample from has infinite tails, you cant use the usual bounding box. You must then choose an enveloping (also called majorizing) function $g(x)$:.

\begin{gathered}
\begin{split}cg(x) \ge  f(x) \qquad\forall x \in (-\infty,\infty)\end{split}\notag\\\begin{split}\end{split}\notag
\end{gathered}

The example below is somewhat comtrived for simplicity.

```python
from scipy.stats import t
```

```python
plt.plot(rainin,gammafunc(rainin, alpha,beta))
plt.plot(rainin,1.4*t.pdf(rainin, loc=alpha-0.25,scale=2.25, df=2));
```

```python
N = 10000 # the total of samples we wish to generate
accepted = 0 # the number of accepted samples
samples = np.zeros(N)
count = 0 # the total count of proposals
p = lambda x: gammafunc(x, alpha,beta)
g = t(loc=alpha-0.25, scale=2.25, df=2)
# generation loop
while (accepted < N):
    

    xproposal = g.rvs()
    
    # pick a uniform number on [0, 1)
    y = np.random.uniform(0,1)
    
    # Do the accept/reject comparison
    if y < p(xproposal)/g.pdf(xproposal):
        samples[accepted] = xproposal
        accepted += 1
    
    count +=1
    
print count, accepted

# get the histogram info
plt.hist(samples,bins=50,  alpha=0.3, normed=True, label=u'Samples');
sns.kdeplot(samples, label="KDE")
plt.plot(rainin,gammafunc(rainin, alpha,beta), label="Original")
plt.legend()
```

Note that these samples are in the inches of rain space, so we have effectively executed a bunch of parametric bootstraps. To doit right, of course, we would want to get as many samples as the original data set.

A quick perusal of fits for the other months tells us that the same gamma wont suffice to model every month, and we will have to indulge ourselves in some hierarchical modelling.

### Importance Sampling for integrals

As you know, another way of doing this calculation is importance sampling. This time we'll use the students-t as the simulation distribution, which we call q.

```python
sz=precip.Dec.shape[0]
sz
```

```python
t_samples = g.rvs(size=sz)
prob_p = p(t_samples)
prob_q = g.pdf(t_samples)
```

```python
w = np.exp(prob_p - prob_q - max(prob_p - prob_q))
w
```
